````python
# prompt: genera il codice di un programma che da in output una tabella di turni di mensa e una di turni stoviglie. gli studenti sono in tutto 35 di cui 17 del secondo anno e 28 del primo. ogni atudente puo al massimo fare un turno mensa e uno stoviglie ogni settimana. ci sono quindi 7 turni a settimana composti da 5 studenti ( 2 del secondo anno e 3 del primo)  . i cognomi degli studenti sono questo : secondi = [
#     'CHIARADIA', 'CONCETTI', 'CORSINOVI', 'CORTICCHIA', 'DI GIROLAMO',
#     'TORCHIA', 'GALLAFIN', 'BORRELLO', 'LUCA', 'MANICONE',
#     'PIERRI', 'ROSSO', 'SPINELLO', 'ZECCHIN', 'RAJARISON',

import random
import pandas as pd

# Dati degli studenti
secondi = [
    'CHIARADIA', 'CONCETTI', 'CORSINOVI', 'CORTICCHIA', 'DI GIROLAMO',
    'TORCHIA', 'GALLAFIN', 'BORRELLO', 'LUCA', 'MANICONE',
    'PIERRI', 'ROSSO', 'SPINELLO', 'ZECCHIN', 'RAJARISON',
    'PORTO', 'MESISCA'
]

primi = [
    'AMORUSO', 'BELLESSO', 'CALDARA', 'CHECCHIA', 'CONSERVA',
    'COSENTINO', 'COSTANTINO', 'DARDANO', 'FERRARA', 'FIORILLO',
    'FONTANELLA', 'GIRALDI', 'IOZZO', 'LANDOLINA', 'OLIVIERI',
    'PAONE', 'PIRRERA', 'PODRINI', 'SAJA', 'SALIERNO',
    'SALVIA', 'SAPONARO', 'SERJANT', 'SPEZZATTI', 'STAIANO',
    'TARAU', 'VENEZIA', 'DI GIOVANNA'
]

# Verifica del numero di studenti
if len(secondi) != 15:
    print("Il numero di studenti del secondo anno non è corretto.")
if len(primi) != 20:
    print("Il numero di studenti del primo anno non è corretto.")


# Funzione per generare i turni
def genera_turni(studenti_secondo, studenti_primo, num_turni, studenti_turno):
    turni = []
    studenti_secondo_disponibili = studenti_secondo.copy()
    studenti_primo_disponibili = studenti_primo.copy()
    for i in range(num_turni):
        turno = []
        num_secondo = 2
        num_primo = 3
        while num_secondo > 0 and studenti_secondo_disponibili:
            studente = random.choice(studenti_secondo_disponibili)
            turno.append(studente)
            studenti_secondo_disponibili.remove(studente)
            num_secondo -= 1
        while num_primo > 0 and studenti_primo_disponibili:
            studente = random.choice(studenti_primo_disponibili)
            turno.append(studente)
            studenti_primo_disponibili.remove(studente)
            num_primo -= 1

        turni.append(turno)
    return turni


# Generazione dei turni mensa
turni_mensa = genera_turni(secondi, primi, 7, 5)

# Generazione dei turni stoviglie
turni_stoviglie = genera_turni(secondi, primi, 7, 5)

# Creazione dei DataFrame
df_mensa = pd.DataFrame(turni_mensa, columns=['Studente 1', 'Studente 2', 'Studente 3', 'Studente 4', 'Studente 5'])
df_stoviglie = pd.DataFrame(turni_stoviglie, columns=['Studente 1', 'Studente 2', 'Studente 3', 'Studente 4', 'Studente 5'])

# Stampa dei DataFrame
print("Turni Mensa:")
print(df_mensa)
print("\nTurni Stoviglie:")
print(df_stoviglie)
````
